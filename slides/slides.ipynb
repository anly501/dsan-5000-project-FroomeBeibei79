## A Plot

```python
echo = False
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Create a plot
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.title('Sine Wave Plot')
plt.grid(True)
plt.show()
```
