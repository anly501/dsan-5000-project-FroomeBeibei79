```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
x = np.linspace(0, 2 * np.pi, 100)  # Create an array of values from 0 to 2*pi
y = np.sin(x)  # Compute the sine of each value

# Create a plot
plt.figure(figsize=(8, 4))  # Set the figure size
plt.plot(x, y, label='sin(x)', color='blue', linestyle='-', linewidth=2)  # Create the plot

# Add labels and a legend
plt.xlabel('x')  # X-axis label
plt.ylabel('y')  # Y-axis label
plt.title('Sine Function')  # Title of the plot
plt.legend()

# Show the plot
plt.grid(True)  # Add grid lines
plt.show()
```
